```python
import random 
```

```python
class play_hangman():
    def __init__(self, level, word, guessed_letters, guessed_words):
        self.level=level
        self.word=word.title()
        self.guessed_letters=guessed_letters.title()
        self.guessed_words=guessed_words.title()
        
    def choose_level():
        level=input("Choose a level to play Hangman! \n1: Easy (Common words with 5 or less letters) \n2: Medium (Common words with 5 or more letters) \n3:Hard (Uncommon words)")
        if level=="1":
            wordbank=["cat","dog","play",'red','blue',"jump","road","tree","bird","car","house",'mice','lamp',"table","chair","mom","dad","can","east",'corn','apple','star','fish','food','meal']
            word=random.choice(wordbank).upper()
            word_completion= " _ " *len(word)
            tries=6
            guessed_letters=[]
            guessed_words=[]
            guessed=False
            print(word_completion)
            while not guessed and tries > 0:
                wordguess = input("Enter a letter or word to guess: ").upper()
                if len(wordguess) == 1 and wordguess.isalpha():
                    if wordguess in guessed_letters:
                        print("You already guessed this letter")
                    elif wordguess not in word:
                        print(wordguess, "is not in the word")
                        tries -= 1
                        guessed_letters.append(wordguess)
                    else:
                        print("Good job!", wordguess, "is in the word!")
                        guessed_letters.append(wordguess)
                        word_as_list= list(word_completion)
                        indices = [i for i, letter in enumerate(word) if letter==wordguess]
                        for index in indices:
                            word_as_list[index] = wordguess
                        word_completion="".join(word_as_list)
                        if word_completion==word:
                            guessed=True
                elif len(wordguess) == len(word) and wordguess.isalpha():
                    if wordguess in guessed_words:
                        print("You already guessed the word", wordguess)
                    elif wordguess != word:
                        print(wordguess, "is not the word")
                        tries -= 1
                        guessed_words.append(wordguess)
                    else:
                        guessed=True
                        word_completion=word
                    
                else:
                    print("That is not a valid guess. Try again.")
                print("You have", tries, "tries left")
                print(word_completion)
                print('\n')
            if guessed:
                print("Congratulations! You won the game!")
            else:
                print("Sorry,you ran out of tries. The word was: ", word)
                
            
        elif level=='2':
            wordbank2=["hornet",'computer','print','purple',"black",'green','trashcan','score','forest','highway','outlet','tablecloth','scorpion','parrot','pencil','cords','mouse','light','sunny','weather','sheet','paper','sense','smell','intelligent']
            word=random.choice(wordbank2).upper()
            word_completion= " _ " *len(word)
            tries=6
            guessed_letters=[]
            guessed_words=[]
            guessed=False
            print(word_completion)
            while not guessed and tries > 0:
                wordguess = input("Enter a letter or word to guess: ").upper()
                if len(wordguess) == 1 and wordguess.isalpha():
                    if wordguess in guessed_letters:
                        print("You already guessed this letter")
                    elif wordguess not in word:
                        print(wordguess, "is not in the word")
                        tries -= 1
                        guessed_letters.append(wordguess)
                    else:
                        print("Good job!", wordguess, "is in the word!")
                        guessed_letters.append(wordguess)
                        word_as_list= list(word_completion)
                        indices = [i for i, letter in enumerate(word) if letter==wordguess]
                        for index in indices:
                            word_as_list[index] = wordguess
                        word_completion="".join(word_as_list)
                        if word_completion==word:
                            guessed=True
                elif len(wordguess) == len(word) and wordguess.isalpha():
                    if wordguess in guessed_words:
                        print("You already guessed the word", wordguess)
                    elif wordguess != word:
                        print(wordguess, "is not the word")
                        tries -= 1
                        guessed_words.append(wordguess)
                    else:
                        guessed=True
                        word_completion=word
                    
                else:
                    print("That is not a valid guess. Try again.")
                print("You have", tries, "tries left")
                print(word_completion)
                print('\n')
            if guessed:
                print("Congratulations! You won the game!")
            else:
                print("Sorry,you ran out of tries. The word was: ", word)
            
        elif level=='3':
            wordbank3=["abyss","sphynx","quartz","absurd",'avenue','awkward','matriarch','calliope','duplicate','sensation','judicial','vow','promise','wander','complex','uncommon','respite','sanctuary','burgeon','bastion','appreciate','conventional']
            word=random.choice(wordbank3).upper()
            word_completion= " _ " *len(word)
            tries=6
            guessed_letters=[]
            guessed_words=[]
            guessed=False
            print(word_completion)
            while not guessed and tries > 0:
                wordguess = input("Enter a letter or word to guess: ").upper()
                if len(wordguess) == 1 and wordguess.isalpha():
                    if wordguess in guessed_letters:
                        print("You already guessed this letter")
                    elif wordguess not in word:
                        print(wordguess, "is not in the word")
                        tries -= 1
                        guessed_letters.append(wordguess)
                    else:
                        print("Good job!", wordguess, "is in the word!")
                        guessed_letters.append(wordguess)
                        word_as_list= list(word_completion)
                        indices = [i for i, letter in enumerate(word) if letter==wordguess]
                        for index in indices:
                            word_as_list[index] = wordguess
                        word_completion="".join(word_as_list)
                        if word_completion==word:
                            guessed=True
                elif len(wordguess) == len(word) and wordguess.isalpha():
                    if wordguess in guessed_words:
                        print("You already guessed the word", wordguess)
                    elif wordguess != word:
                        print(wordguess, "is not the word")
                        tries -= 1
                        guessed_words.append(wordguess)
                    else:
                        guessed=True
                        word_completion=word
                    
                else:
                    print("That is not a valid guess. Try again.")
                print("You have", tries, "tries left")
                print(word_completion)
                print('\n')
            if guessed:
                print("Congratulations! You won the game!")
            else:
                print("Sorry,you ran out of tries. The word was: ", word)
        
```
